```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import (Input, Embedding, LSTM, Dense, Dropout,
                                     Bidirectional, concatenate)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
import pickle, os
```

```python
# --- 2. Load Dataset
df = pd.read_csv('/content/drive/MyDrive/sem7 project ibm/Finance_Trends.csv')
print("✅ Data loaded successfully. Shape:", df.shape)
display(df.head())
```

```python
# --- 3. Combine Text Columns
text_cols = [c for c in df.columns if any(k in c.lower()
                                          for k in ['reason','objective','purpose','expect','source','comment','feedback'])]
df['combined_text'] = df[text_cols].fillna('').astype(str).agg(' . '.join, axis=1)
```

```python
# --- 4. Identify Numeric Investment Columns
invest_cols = [c for c in df.columns
               if any(k in c.lower()
                      for k in ['mutual','equity','bond','deposit','gold','ppf','share','fund'])]

print("Text columns used:", text_cols)
print("Investment columns used:", invest_cols)
```

```python
# --- 5. Create Labels (Risk Personality)
# Assign heuristic weights
weights = {}
for c in invest_cols:
    lc = c.lower()
    if 'equity' in lc or 'share' in lc:
        weights[c] = 1.2
    elif 'mutual' in lc:
        weights[c] = 1.0
    elif 'bond' in lc or 'debenture' in lc:
        weights[c] = 0.8
    elif 'gold' in lc:
        weights[c] = 0.6
    else:  # deposits, ppf etc.
        weights[c] = -0.8

# ✅ Filter out non-numeric columns (the cause of the error)
numeric_invest_cols = [c for c in invest_cols if pd.api.types.is_numeric_dtype(df[c])]

# If no numeric columns found, select numeric ones from all columns as fallback
if len(numeric_invest_cols) == 0:
    numeric_invest_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print("Numeric investment columns used:", numeric_invest_cols)

# Fill NA and compute risk score only on numeric columns
df[numeric_invest_cols] = df[numeric_invest_cols].fillna(df[numeric_invest_cols].mean())

# Use only numeric columns in the weighted sum
df['risk_score'] = sum(df[c] * weights.get(c, 0) for c in numeric_invest_cols)

# 3 equal bins for Low / Medium / High risk
df['risk_label'] = pd.qcut(df['risk_score'], q=3, labels=['Low', 'Medium', 'High'])
label_map = {'Low': 'Safe Investor', 'Medium': 'Balanced Planner', 'High': 'Risk Taker'}
df['personality'] = df['risk_label'].map(label_map)

sns.countplot(x='personality', data=df)
plt.title("Distribution of Personality Classes")
plt.show()
```

```python
# --- Encode Target Labels (Add this immediately after Step 5) ---
from sklearn.preprocessing import LabelEncoder

target_le = LabelEncoder()
df['target_encoded'] = target_le.fit_transform(df['personality'])

# Define target variable for model training
y = df['target_encoded']

print("✅ Encoded target labels:", list(target_le.classes_))
```

```python
# --- 6. Text Tokenization (Updated & Safe)

# Identify text columns again (including textual investment reasons/objectives)
text_cols = [c for c in df.columns if df[c].dtype == 'object' and c != 'personality']

# Combine all text columns into a single string per record
df['combined_text'] = df[text_cols].fillna('').astype(str).agg(' . '.join, axis=1)

# Optional: Clean excessive spaces or dots
df['combined_text'] = df['combined_text'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Check a sample
print("✅ Combined text sample:")
print(df['combined_text'].head(2).values)

# Tokenization parameters
MAX_NUM_WORDS = 30000
MAX_SEQ_LEN = 150

# Create and fit tokenizer
tokenizer = Tokenizer(num_words=MAX_NUM_WORDS, oov_token='<OOV>')
tokenizer.fit_on_texts(df['combined_text'])

# Convert texts to integer sequences
seqs = tokenizer.texts_to_sequences(df['combined_text'])

# Pad sequences to fixed length
padded = pad_sequences(seqs, maxlen=MAX_SEQ_LEN, padding='post', truncating='post')

print("✅ Text tokenization complete.")
print("Vocabulary size:", len(tokenizer.word_index))
print("Shape of padded sequences:", padded.shape)
```

```python
# --- 7. Numeric Feature Processing (Updated & Safe)

from sklearn.preprocessing import LabelEncoder, StandardScaler

num_df = pd.DataFrame()

# --- Age ---
if 'age' in df.columns:
    num_df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(df['age'].median())
else:
    num_df['age'] = 30  # default

# --- Gender ---
if 'gender' in df.columns:
    df['gender'] = df['gender'].fillna('Unknown').astype(str)
    gle = LabelEncoder()
    num_df['gender_enc'] = gle.fit_transform(df['gender'])
else:
    num_df['gender_enc'] = 0

# --- Investment Columns ---
numeric_invest_cols = []
encoded_invest_cols = []

for c in invest_cols:
    if pd.api.types.is_numeric_dtype(df[c]):
        # numeric column — keep as-is
        num_df[c] = df[c].fillna(df[c].mean())
        numeric_invest_cols.append(c)
    else:
        # non-numeric (text like 'Dividend', 'Liquidity', etc.)
        le = LabelEncoder()
        df[c] = df[c].fillna('Unknown').astype(str)
        num_df[c + '_enc'] = le.fit_transform(df[c])
        encoded_invest_cols.append(c + '_enc')

# --- Scale all numeric features ---
scaler = StandardScaler()
num_scaled = scaler.fit_transform(num_df)

print("✅ Numeric feature processing complete.")
print("Numeric columns used:", numeric_invest_cols)
print("Encoded text columns:", encoded_invest_cols)
print("Shape of numeric feature matrix:", num_scaled.shape)
```

```python
# --- 8. Labels Encoding
le = LabelEncoder()
y = le.fit_transform(df['personality'])
```

```python
# --- 9. Train-Test Split
X_text_train, X_text_test, X_num_train, X_num_test, y_train, y_test = train_test_split(
    padded, num_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# --- 10. Build Deep Learning Model
text_in = Input(shape=(MAX_SEQ_LEN,), name='text_input')
embed = Embedding(input_dim=MAX_NUM_WORDS, output_dim=128, input_length=MAX_SEQ_LEN)(text_in)
x = Bidirectional(LSTM(128, dropout=0.3, recurrent_dropout=0.3))(embed)
x = Dense(64, activation='relu')(x)
x = Dropout(0.3)(x)

num_in = Input(shape=(X_num_train.shape[1],), name='num_input')
n = Dense(64, activation='relu')(num_in)
n = Dropout(0.2)(n)
n = Dense(32, activation='relu')(n)

combined = concatenate([x, n])
z = Dense(64, activation='relu')(combined)
z = Dropout(0.3)(z)
out = Dense(len(le.classes_), activation='softmax')(z)

model = Model([text_in, num_in], out)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# --- 11. Train Model
es = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
history = model.fit([X_text_train, X_num_train], y_train,
                    validation_split=0.1, epochs=10, batch_size=64,
                    callbacks=[es], verbose=1)
```

```python
# --- 12. Evaluate
loss, acc = model.evaluate([X_text_test, X_num_test], y_test, verbose=0)
print(f"\n✅ Test Accuracy: {acc*100:.2f}%")

y_pred = np.argmax(model.predict([X_text_test, X_num_test]), axis=1)
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# --- 13. Save Artifacts ---
import os, pickle

save_dir = "/content/drive/MyDrive/sem7 project ibm/finance_personality_artifacts"
os.makedirs(save_dir, exist_ok=True)

# --- Save model ---
model.save(f"{save_dir}/finance_personality_model.keras")

# --- Save tokenizer, scaler, and encoders ---
with open(f"{save_dir}/tokenizer.pkl", "wb") as f:
    pickle.dump(tokenizer, f)

with open(f"{save_dir}/scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

with open(f"{save_dir}/label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

with open(f"{save_dir}/target_label_encoder.pkl", "wb") as f:
    pickle.dump(target_le, f)

print(f"📁 All model artifacts saved in {save_dir}")
```

```python
# --- 14. Example Prediction (Safe & Flexible) ---
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Example user input
sample_text = ["I invest mostly in fixed deposits and gold for safety"]

# Create a DataFrame with the same numeric structure used for training
sample_num = pd.DataFrame(columns=num_df.columns)

# Fill in values for known features
sample_num.loc[0, 'age'] = 35
sample_num.loc[0, 'gender_enc'] = 0  # assume 0 for 'Male' if encoded that way

# Fill investment columns with some example numeric/text-like patterns
for c in sample_num.columns:
    if c not in ['age', 'gender_enc']:
        sample_num.loc[0, c] = np.random.uniform(1000, 10000)

# Apply same scaling as training
sample_invest_scaled = scaler.transform(sample_num)

# Process text
sample_seq = tokenizer.texts_to_sequences(sample_text)
sample_pad = pad_sequences(sample_seq, maxlen=MAX_SEQ_LEN, padding='post', truncating='post')

# Predict
pred = model.predict([sample_pad, sample_invest_scaled])

# Get predicted label (target_le is used for target encoding)
pred_label = target_le.inverse_transform([np.argmax(pred)])

print("\n💡 Predicted Personality:", pred_label[0])
```
